```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


print("Loading US Accidents dataset...")
file_path = 'US_Accidents_March23.csv'
us_accidents = pd.read_csv(file_path)
len_of_df = len(us_accidents)
print(f"Sample size: {len(us_accidents)}")
```

```python
us_accidents.head()
```

```python
us_accidents["End_Time"] = pd.to_datetime(us_accidents["End_Time"], errors="coerce", format="%Y-%m-%d %H:%M:%S.%f")

us_accidents["End_Hour"] = us_accidents["End_Time"].dt.hour
```

```python
plt.figure(figsize=(10, 6))
us_accidents['End_Hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Accidents by Year')
plt.xlabel('End Time (Hours)')
plt.ylabel('Number of Accidents')
plt.show()
```
